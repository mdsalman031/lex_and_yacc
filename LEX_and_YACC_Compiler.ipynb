## Lex only

```python
#@title Writing Lex program
%%writefile program.l

%{
#include<stdio.h>
int vowels = 0;
int consonants = 0;
%}

%%
[aeiouAEIOU] {vowels++;}
[a-zA-Z] {consonants++;}
\n {printf("Vowels - %d, consonants - %d", vowels, consonants);}
%%

int yywrap() {
  return 1;
}

int main() {
  printf("Enter a string :");
  yylex();
  return 0;
}
```

```python
%%writefile program.l

%{
  #include<stdio.h>
%}

%%
[a-z] {printf("%c", yytext[0] - 32);}
. {printf("%s", yytext);}
%%

int yywrap() {
  return 1;
}

int main() {
  printf("Enter a string : ");
  yylex();
  return 0;
}
```

## Lex and Yacc combined

```python
#@title Writing YACC program
%%writefile program.y

%{
    #include<stdio.h>
    #include<stdlib.h>
%}
%token DIGIT LETTER UND NL
%%
stmt: variable NL {printf("Valid Identifier\n");exit(0);}
variable: LETTER alphanumeric;
alphanumeric: LETTER alphanumeric
            | DIGIT alphanumeric
            | UND alphanumeric
            | LETTER
            | DIGIT
            | UND;
%%

int yyerror(){
    printf("Invalid Identifier\n");
    exit(0);
}

void main(){
    printf("Enter the variable name: ");
    yyparse();
}
```

```python
#@title Writing Lex program
%%writefile program.l

%{
    #include "y.tab.h"
%}
%%
[a-zA-Z] {return LETTER;}
[0-9] {return DIGIT;}
[_] {return UND;}
\n {return NL;}
. {return yytext[0];}
%%
```
