## Lex only

```python
#@title Writing Lex program
%%writefile program.l

%{
#include<stdio.h>
int vowels = 0;
int consonants = 0;
%}

%%
[aeiouAEIOU] {vowels++;}
[a-zA-Z] {consonants++;}
\n {printf("Vowels - %d, consonants - %d", vowels, consonants);}
%%

int yywrap() {
  return 1;
}

int main() {
  printf("Enter a string :");
  yylex();
  return 0;
}
```

```python
%%writefile program.l

%{
  #include<stdio.h>
%}

%%
[a-z] {printf("%c", yytext[0] - 32);}
. {printf("%s", yytext);}
%%

int yywrap() {
  return 1;
}

int main() {
  printf("Enter a string : ");
  yylex();
  return 0;
}
```

```python
%%writefile program.l

%{
    #include<stdio.h>
%}

%%
0[xX][0-9a-fA-F]+ {printf("Hexadeximal number : %s\n", yytext);}
0[0-7]+ printf("Octal number : %s\n", yytext);
%%

int yywrap() {
    return 1;
}

int main() {
  printf("Enter a number : ");
  yylex();
  return 0;
}
```

```python
%%writefile program.l

%{
#include<stdio.h>
%}

%%
-[0-9]+ {printf("Negative number : %s\n", yytext);}
[0-9]+ {printf("Positive number : %s\n", yytext);}
%%

int yywrap() {
  return 1;
}

int main() {
  printf("Enter a number : ");
  yylex();
  return 0;
}
```

```python
%%writefile program.l

%{
#include<stdio.h>
%}

%%
int|if|include {printf("Keyword starting with i : %s\n", yytext);}
i[a-zA-Z0-9_]+ {printf("Identifier starting with i : %s\n", yytext);}
%%

int yywrap() {
  return 1;
}

int main() {
  printf("Enter a string : ");
  yylex();
  return 0;
}
```

## Lex and Yacc combined

```python
#@title Writing YACC program
%%writefile program.y

%{
  #include<stdio.h>
  #include<stdlib.h>

  void yyerror(char *s);
  int yylex();
%}

%left '+' '-'
%left '*' '/'

%token NUMBER

%%
input:
      | input line
      ;
line:
      expr '\n' {printf("Result : %d\n", $1);}
      ;
expr:
      expr '+' expr {$$ = $1 + $3;}
      | expr '-' expr {$$ = $1 - $3;}
      | expr '*' expr {$$ = $1 * $3;}
      | expr '/' expr {$$ = $1 / $3;}
      | '(' expr ')' {$$ = $2;}
      | NUMBER {$$ = $1;}
      ;
%%

int main() {
  printf("Enter expression : \n");
  yyparse();
  return 0;
}

void yyerror(char *s) {
  printf("Error : %s", s);
}
```
